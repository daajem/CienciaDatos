```python
from sklearn import datasets
import pandas as pd
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
print(dataset)
```

```python
print('Informacion del dataset')
print(dataset.keys())
print()
```

```python
print('Caracteristicas')
print(dataset.Age)
print()
print(dataset.EstimatedSalary)
print()
print(dataset.Purchased)
#Lo puse asi por que segui el video, y por que el 'DESCR' me daba error me imagino por algun elemento del dataset que mostraba de ejemplo en el video
```

```python
X = dataset[['Age', 'EstimatedSalary']]
y = dataset['Purchased']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
algoritmo = DecisionTreeClassifier(criterion = 'entropy')
```

```python
algoritmo.fit(X_train, y_train)
```

```python
y_pred = algoritmo.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(y_test, y_pred)
print('Matriz')
print(matriz)
```

```python
from sklearn.metrics import precision_score
precision = precision_score(y_test, y_pred)
print('Precision Modelo')
print(precision)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(12, 6))
plot_tree(algoritmo, feature_names=['Age', 'EstimatedSalary'], class_names=['No Compra', 'Compra'], filled=True)
plt.show()
```
