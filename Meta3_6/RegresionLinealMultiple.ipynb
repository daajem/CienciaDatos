```python
import numpy as np
from sklearn import linear_model
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt
```

```python
california = fetch_california_housing()
print(california)
```

```python
print('Informacion en el dataset:')
print(california.keys())
print()
```

```python
print('Caracteristicas en el dataset:')
print(california.DESCR)
print()
```

```python
print('Cantidad de datos')
print(california.data.shape)
```

```python
X_multiple = california.data[:, 5:8]
print(X_multiple)
```

```python
y_multiple = california.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_multiple, y_multiple, test_size=0.2)
```

```python
lr_multiple = linear_model.LinearRegression()
```

```python
lr_multiple.fit(X_train, y_train)
```

```python
Y_pred_multiple = lr_multiple.predict(X_test)
```

```python
print('DATOS DEL MODELO')
print('Valor de las pendientes p coeficiente "a"')
print(lr_multiple.coef_)
print('Valos de la interseccion o coeficiente "b"')
print(lr_multiple.intercept_)
```

```python
print('Precision del modelo')
print(lr_multiple.score(X_train, y_train))
```
